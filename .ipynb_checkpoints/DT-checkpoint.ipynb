```python
Dataset = pd.read_csv("Bank_dataset.csv",sep=';')
Dataset_Copy= Dataset.copy()
```

# Binning Age column

```python
AgeBin=[]
for age in (Dataset["age"].values):
    if (age >0 and age <=16):
        AgeBin.append("Child")
    elif (age >16 and age <=30):
        AgeBin.append("Young Adult")
    elif (age >30 and age <=45):
        AgeBin.append("Middle-Aged Adult")
    elif (age >45):
        AgeBin.append("Old-Aged Adult")
```

# Change UNKNOWN Value to UNKNOWN + COLNAME

```python
for (colname,colval) in Dataset_Copy.items():
    Dataset_Copy.loc[(Dataset_Copy[colname] == 'unknown'),colname] ='unknown'+colname
```

```python
Dataset_Copy.describe()
```

```python
Dataset_Copy.insert(1, 'Age Group', AgeBin)
Dataset_Copy
```

# Spliting Data

```python
SplitPercentage= float(input("Please Enter Percentage of Train Data: "))
```

```python
# the X features
x = Dataset_Copy[["Age Group","job","marital","education","housing"]].values

# Output Value
y= Dataset_Copy["y"].values   
```

```python
#Shuffle Dataset and split
from sklearn.model_selection import train_test_split
x_Train, x_Test, y_Train, y_Test = train_test_split(x, y, test_size=(1-SplitPercentage), random_state=42)
print(x_Train.shape)
print(x_Test.shape)
print(y_Train.shape)
print(y_Test.shape)
```

# DECISION TREE ALGORITHM

```python
class Node: # Represent the feature column
    def __init__(self,data,output):
        self.data = data
        
        self.children = {}
        
        self.output = output # Output is empty if tree is empty
        
    def add_child(self,Value,obj):
        self.children[Value] = obj 
        #EACH CHILD Carry the value of the feature and tha node for its split
```

```python
class DecisionTree:
    def __init__(self):
        self.root = None
        
    def GetClassesCount(self,y_train): # Returns classes and its count in data
        Class_dict = dict() 
        Y_set= set(y_train) # yes ,no
        for Class in Y_set:
            Class_dict[Class] = list(y_train).count(Class) #-->yes :count
        return Class_dict
        
    def CalculateEntropy(self,y_train): # Calculate ENTROPY
        Classes_dict= self.GetClassesCount(y_train) #-->yes :count
        Entropy = 0
        for Class in Classes_dict: 
            p = Classes_dict[Class] / len(y_train)
            Entropy += (-p) * math.log2(p)    #-p'yes'*log2(p'yes')-p'no'*log2(p'no')
        return Entropy
    
    def GainRatio(self,x_train,y_train,Index_FeatureSelected): #calculating Gain Ratio 'Information Gain for Each Feature Selected'
        Entropy_Column = self.CalculateEntropy(y_train) #Entropy for whole Data --> yes ,no
        
        Entropy_Feature = 0 # after Spliting Feature Node
        
        split_info = 0
        
        Feature_Values = set(x_train[:,Index_FeatureSelected]) #--> Unique feature Values
        TrainDataset = pd.DataFrame(x_train)
        featurescount = TrainDataset.shape[1]
        TrainDataset[featurescount]=y_train
        TrainingSize=TrainDataset.shape[0]
        for i in Feature_Values:
            FeatureRecords = TrainDataset[TrainDataset[Index_FeatureSelected] == i]
            size=FeatureRecords.shape[0] #  for example, how many married in Training Data
            p = size / TrainingSize
            Entropy_Feature+= (p) * self.CalculateEntropy(FeatureRecords[FeatureRecords.shape[1]-1])
            split_info += (-p) * math.log2(p)
        if split_info == 0 :
            return math.inf
        FeatureInformationGain = Entropy_Column - Entropy_Feature
        FeatureGainRatio = FeatureInformationGain / split_info
        return FeatureGainRatio
    
    def DecisionTree(self,x_train,y_train,Features,Depth_level,classes):
        
        if len(classes) == 1:
            self.output = None
            for i in classes:
                if i in y_train:
                    output = i
            return Node(None,output)
        
        Classes_dict= self.GetClassesCount(y_train)
        if len(Features)==0: # contains Features index
            output = None
            max_count = -100000000 
            for i in classes:
                if i  in Classes_dict:
                    if Classes_dict[i] > max_count :
                        output = i
                        max_count = Classes_dict[i]
            return Node(None,output)
        
        
        # Finding the best feature to split upon 
        Max_Information_Gain = -100000000 # -Infinity
        FeatureSplitted_MaxGain = None
        
        for f in Features :
            current_gain = self.GainRatio(x_train,y_train,f)
            if current_gain > Max_Information_Gain:
                Max_Information_Gain = current_gain
                FeatureSplitted_MaxGain = f 
                
        output = None
        max_count = -100000000 
        for i in classes:
            if i  in Classes_dict:
                if Classes_dict[i] > max_count :
                    output = i
                    max_count = Classes_dict[i]
        FeatureUniqueValues = set(x_train[:,FeatureSplitted_MaxGain])  
        Dataframe= pd.DataFrame(x_train)
        Dataframe[Dataframe.shape[1]] = y_train
        
        current_node = Node(FeatureSplitted_MaxGain,output)
        index  = Features.index(FeatureSplitted_MaxGain)
        Features.remove(FeatureSplitted_MaxGain)
        for unique in FeatureUniqueValues:
            RemainingData = Dataframe[Dataframe[FeatureSplitted_MaxGain] == unique]
            #---------------> Building Tree Recursively
            classes = set(RemainingData.iloc[:,RemainingData.shape[1]-1].values)
            # change classes for Purity
            node = self.DecisionTree(RemainingData.iloc[:,0:RemainingData.shape[1]-1].values,
                                        RemainingData.iloc[:,RemainingData.shape[1]-1].values,
                                          Features,Depth_level+1,classes)
            current_node.add_child(unique,node)
        Features.insert(index,FeatureSplitted_MaxGain)
        return current_node
    
    def fit(self,x_train,y_train):
        
        Features = [i for i in range(x_train.shape[1])]
        Classes = set(y_train)
        level = 0
        
        self.root = self.DecisionTree(x_train,y_train,Features,level,Classes)
        
    def predictRecursively(self,Datatestpoint,RootNode): #------> Recursively
        if len(RootNode.children) == 0 :
            return RootNode.output

        value = Datatestpoint[RootNode.data]
        if value not in RootNode.children :
            return RootNode.output
        return self.predictRecursively(Datatestpoint,RootNode.children[value])
    
    def predict(self,x_test):
        
        Y_Predictions= np.array(['NAN' for i in range(len(x_test))])
        for i in range (x_test.shape[0]):
            Y_Predictions[i]= self.predictRecursively(x_test[i],self.root)
        
        return Y_Predictions
    
    def Accuracy(self,y_test,ypred): # Calculate the Accuracy for the Model
        
        rightPredictions= np.sum(ypred==y_test)
        return rightPredictions/len(y_test)*100
    
```

```python
d=DecisionTree()
d.fit(x_Train,y_Train)
```

```python
y_predict = d.predict(x_Test)
AccuracyDT=d.Accuracy(y_Test,y_predict)
print(AccuracyDT,'%')
```

```python
tr = pd.DataFrame(data=y_Test, columns=['Loan acceptability'])
pr = pd.DataFrame(data=y_predict, columns=['Loan acceptability'])


fig, ax = plt.subplots(1, 2, sharex='col', sharey='row', figsize=(15,6))


sns.countplot(x='Loan acceptability', data=tr, ax=ax[0], palette='viridis', alpha=0.7)
sns.countplot(x='Loan acceptability', data=pr, ax=ax[1], palette='viridis', alpha=0.7)


fig.suptitle('Decision Tree Model', fontsize=20)

ax[0].tick_params(labelsize=12)
ax[1].tick_params(labelsize=12)
ax[0].set_title("True Data", fontsize=18)
ax[1].set_title("Predicted Data", fontsize=18)
plt.show()
```
